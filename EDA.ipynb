```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Lokesh\Desktop\House_Price_Prediction\Bengaluru_House_Data.csv.xls')
```

```python
df
```

```python
df = df.drop(columns=['area_type','availability','society','balcony'])
```

```python
df.isnull().sum()
```

```python
df['location'].value_counts()
```

```python
df['location'] = df['location'].fillna('Sarjapur  Road')
```

```python
df.isnull().sum()
```

```python
df['size'].value_counts()
```

```python
df['size'] = df['size'].fillna('2 BHK')
```

```python
df.isnull().sum()
```

```python
df['bath'].value_counts()
```

```python
df['bath'] = df['bath'].fillna(2.0).astype(int)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df['location'] = df['location'].apply(lambda x: x.strip())
```

```python
loc = df['location'].value_counts()
loc_less_than_10 = loc[loc <= 10]

df['location'] = df['location'].apply(lambda x: 'others' if x in loc_less_than_10.index else x)
```

```python
out = [int(i.split()[0]) for i in df['size']]
df['bhk'] = out
df
```

```python
def clean_sqft(sqft):
    l = sqft.split("-")
    if len(l) == 2:
        return float(l[0]) + float(l[1]) / 2
    try:
        return float(l[0])
    except:
        return None
df['total_sqft'] = df['total_sqft'].apply(clean_sqft)
```

```python
df
```

```python
df['total_sqft'] = df['total_sqft'].fillna(round(df['total_sqft'].mean()))
df
```

```python
df['price_per_sqft'] = df['price']* 100000 / df['total_sqft']
df
```

```python
df['total_sqft'].describe()
```

```python
df = df[df["total_sqft"]/df['bhk'] >= 300]
df['total_sqft'].describe()
```

```python
sns.boxplot(x='price_per_sqft',data=df)
```

```python
q1=df["price_per_sqft"].quantile(0.25)#0-0.25 0.25-0.5 0.5-0.75 0.75-1
q3=df["price_per_sqft"].quantile(0.75)

IQR = q3 - q1

lower=q1-0.5*IQR
upper=q3+0.5*IQR
```

```python
df=df[(df["price_per_sqft"]>=lower) & (df["price_per_sqft"]<=upper)]
```

```python
sns.boxplot(x="price_per_sqft",data=df)
```

```python
df.reset_index(inplace=True)
df
```

```python
df=df.drop(["index","price_per_sqft","size"],axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder,Normalizer,StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
encoder = LabelEncoder()
df['encoded_loc'] = encoder.fit_transform(df['location'])
df
```

```python
x = df.drop(['location','price'],axis=1)
y = df.price
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = RandomForestRegressor(random_state=42)
```

```python
params = {
    "n_estimators":[100,150,200,250,300],
    "max_depth":[3,4,5,6,7]
}
```

```python
grid = GridSearchCV(estimator=model,param_grid=params,cv=5)
grid.fit(x_train,y_train)
```

```python
y_pred = grid.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
grid.score(x_train,y_train)
```

```python
grid.score(x_test,y_test)
```
